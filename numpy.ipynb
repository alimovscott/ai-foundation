```python
import numpy as np
```

```python
# 1D array

import numpy as np

arr = np.array([1, 3, 5, 7])
print('numpy_array:', arr)
```

```python
# 2D array -> matrix
arr2 = np.array([[1, 3, 5, 7], [4, 6, 8, 9]])
print('numpy_array2:', arr2)

arr2.shape
```

```python
# dtype
a = np.array([1, 2, 3])
print(a.dtype)
```

```python
# reshape()

a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape(2, 3)

print("shaped_array:", b)
```

```python
#  astype()

a = np.array([1, 2, 3])
b = a.astype(float)

print("float:", b)
```

```python
# np.zeros(), np.ones() => dummy data

z = np.zeros((2, 3))
o = np.ones((2, 3))

print("zeros:", z)
print("ones:", o)
```

```python
# ndim => used to find dimension

a = np.array([[1, 2], [3, 4]])
print("ndim:", a.ndim)
```

```python
# arange() => used to create array for test data

a = np.arange(1, 6)

print("arange:", a)
```

```python
# sum(), mean(), max(), min()

a = np.array([1, 2, 3, 4])

print("sum: ", a.sum())
print("mean: ", a.mean())
print("max: ", a.max())
print("min: ", a.min())
```

```python
# flatten() => 2D/3D -> into 1D array

a = np.array([[1, 2], [3, 4]])
b = a.flatten()

print("flatten:", b)
```

```python
# T (Transpose)

a = np.array([[1, 2], [3, 4]])
print("transpose: ", a.T)
```

```python
# dot() - @ (Matrix maltiplication)

a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 0], [1, 2]])

# print("dot/maltiplication: ", a.dot(b))
print("dot/maltiplication: ", a @ b)
```

```python
# random.rand() - creates random numbers between [0, 1)

a = np.random.rand(2, 3)
print("random: ", a)
```

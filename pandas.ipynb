```python

# creating a dataset with missing data

data_with_missing = {
    "cars": ["BMW", "Volvo", None, "Tesla", "Ferrari", None, "BMW"],
    "counts": [3, 7, None, 8, 1, 9, None]
}

df_missing = pd.DataFrame(data_with_missing)
print(df_missing)
```

```python
# counting missing data

df_missing.isnull().sum()
```

```python

# dropna() - deleting missing data

df_missing.dropna()
df_missing.dropna(how="all")  # checks by rows with all null data
df_missing.dropna(axis=1)     # checks by column
df_missing.dropna(subset=["counts"]) # checks by subsets
```

```python

# fillna() - filling missing data

df_missing.fillna("unknown")
df_missing.fillna(0)

df_missing.fillna({"cars": "unknown", "counts": 0})
```
